```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from datetime import date
import matplotlib.axes as ax
from sklearn.linear_model import LinearRegression
```

```python
v_meter = pd.read_csv(r"http://api.covid19india.org/csv/latest/vaccine_doses_statewise_v2.csv")
v_meter
```

```python
v_meter.isnull().sum()
```

```python
x = v_meter.iloc[6269]['First Dose Administered']
y = v_meter.iloc[6269]['Second Dose Administered']
z = x-y
vaccine_num = (y, z)
plot_labels = ('Fully Vaccinated', 'Partially Vaccinated')
ex_plode = (0.2 , 0)
kolor = ('#5EB0E5', '#7F8B52')
```

```python
plt.figure(figsize =(5 , 5))
plt.pie(vaccine_num , explode = ex_plode , shadow = True , colors = kolor , labels =plot_labels , autopct='%1.2f%%')
plt.legend()
```

```python
data1 = pd.read_csv(r"http://api.covid19india.org/csv/latest/cowin_vaccine_data_statewise.csv")
data1
```

```python
male = data1.iloc[167]['Male(Individuals Vaccinated)']
female = data1.iloc[167]['Female(Individuals Vaccinated)']
vaccine_gen = (male, female)
plot_la_bels = ('MALE', 'FEMALE')
ex_plode = (0.05 , 0)
kolor = ('#C46210', '#BC5090')
```

```python
plt.figure(figsize =(5 , 5))
plt.pie(vaccine_gen , explode = ex_plode , shadow = True , colors = kolor , labels =plot_la_bels , autopct='%1.2f%%')
plt.legend()
```

```python
group1 = data1.iloc[167]['18-45 years (Age)']
group2 = data1.iloc[167]['45-60 years (Age)']
group3 = data1.iloc[167]['60+ years (Age)']
x_axis = [group1 , group2 , group3]
y_axis = ['18-45 years (Age)' , '45-60 years (Age)' , '60+ years (Age)']
plt.figure(figsize=(7 , 7))
plt.bar(y_axis , x_axis , width = 0.5 )
plt.title('Age Group Vaccination Details')
plt.gcf().axes[0].yaxis.get_major_formatter().set_scientific(False)
plt.show()
```

```python
data2 = data1[data1['Updated On'] == '30/06/2021']
data2 = data2[1:]
data2 = data2.sort_values('Total Individuals Vaccinated' , ascending=False )
data2
```

```python
states = data2['State']
Total = data2['Total Individuals Vaccinated']
plt.figure(figsize=(20 , 10))
plt.bar( states , Total)
plt.title('Statewise Vaccination Details')
plt.gcf().axes[0].yaxis.get_major_formatter().set_scientific(False)
plt.xticks(rotation=60)
plt.show()
```

```python
gogo = data1[data1['State'] == 'India']
gogo = gogo.iloc[:-2]
gogo = gogo.iloc[::7]
gogo
```

```python
Vaccinated_As_of = gogo['Updated On']
Total_Individuals_Vaccinated = gogo['Total Individuals Vaccinated']
First_Dose_Administered = gogo['First Dose Administered']
Second_Dose_Administered = gogo['Second Dose Administered']
plt.figure(figsize=(15 , 10))
plt.plot(Vaccinated_As_of , Total_Individuals_Vaccinated , marker='o' , label = 'First Dose Administered'  )
plt.title('Vaccination Trends')
plt.plot(Vaccinated_As_of , Second_Dose_Administered , marker='o' , label = 'Second Dose Administered '  )
plt.ticklabel_format(style='sci', axis='y', scilimits=(-10000000,100000000))
plt.xticks(rotation=60)
plt.legend()
plt.show()
```

```python
radar = gogo['Total Individuals Vaccinated'].tolist()
radar = list(map(int , radar))
radar.pop(0)
```

```python
#calculating weekly average vaccinations
for i in range(23):
    radar[i] = (radar[i+1]-radar[i])
radar1 = radar[:-1]
weekly_mean = sum(radar1)/len(radar1)
int(weekly_mean)
```
